```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
data = pd.read_csv("clustered_forestfires.csv")
data.loc[:, "log_area"] = np.log(data["area"] + 1)
```

```python
data.loc[data["area"] <= 2.14, "area_type"] = 1
data.loc[(data["area"] > 2.14) & (data["area"] <= 15.42), "area_type"] = 2
data.loc[data["area"] > 15.42, "area_type"] = 3
```

```python
data.loc[:, "month_type"] = 1
data.loc[(data["month"] == "sep") | (data["month"] == "aug"), "month_type"] = 2
```

```python
data.loc[:, "temp_type"] = 1
data.loc[(data["temp"] >= 15) & (data["temp"] <= 25), "temp_type"] = 2
```

```python
data.loc[:, "RH_type"] = 1
data.loc[(data["RH"] >= 20) & (data["RH"] <= 60), "RH_type"] = 2
```
